```r
suppressMessages(library("DESeq2"))
library("ggplot2")
library("RColorBrewer")
```

### RBFOX2 50ng 24hr vs RBFOX0 ###

```r
counts <- read.table('/home/rmarina/projects/apobec/sample_featurecounts/output/RBFOX_v_RB0_24hr_50_counts.txt', header=TRUE, row.names=1)
conditions <- read.table('/home/rmarina/projects/apobec/sample_featurecounts/output/RBFOX_v_RB0_24hr_50_conditions.txt', header=TRUE, row.names=1)
conditions
```

```r
dds <- DESeqDataSetFromMatrix(countData = counts,
                              colData = conditions,
                              design = ~ conditions)
```

```r
dds <- DESeq(dds)
```

```r
res <- results(dds,lfcThreshold = 0,alpha=0.05)
summary(res)
```

```r
plotMA(res, main="DESeq2", ylim=c(-7,7))
```

```r
write.csv(as.data.frame(res), file="/home/rmarina/projects/apobec/sample_featurecounts/output/deseq/rbfox2_vs_RB0_24_50_deseq.csv")
```
