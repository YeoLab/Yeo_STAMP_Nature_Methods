```python
def load_eclip(eclip_filepath):
    """Load the eCLIP data at the specified filepath. 
    Columns must be restricted to the following, and ordered as follows:
    'chrom','start','end','l10p','l2fc','strand', 'region', 'geneid'
    """
    eclip = pd.read_csv(eclip_filepath, sep='\t')
    num_columns = len(eclip.columns)
    expected_columns = ['chrom','start','end','l10p','l2fc','strand', 'region', 'geneid']
    if num_columns > len(expected_columns):
        num_dummy_cols_to_add = num_columns - len(expected_columns)
        columns_to_add = expected_columns + list(['{}_dummy'.format(d) for d in range(0,num_dummy_cols_to_add)])
    else:
        columns_to_add = expected_columns
    eclip.columns = columns_to_add
    
    return eclip
```

```python
from collections import defaultdict


def get_distance_from_stamp_site(r):
    abs_to_real = defaultdict()
    
    try:
        positions = [int(p) for p in r.GCATG_pos.split(',')]
        dists_from_center = [p - 200 for p in positions]
        for d in dists_from_center:
            abs_to_real[abs(d)] = d

        return abs_to_real.get(min(abs_to_real.keys()))
    except:
        return None

```
