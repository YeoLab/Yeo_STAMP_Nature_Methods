```python
import pandas as pd
import numpy as np
import glob
import os
from tqdm import tnrange, tqdm_notebook
```

```python
input_dir = '/home/bay001/projects/kris_apobec_20200121/permanent_data2/07_scRNA_groups/window_outputs_deep_cds_only'
output_dir = '/home/bay001/projects/kris_apobec_20200121/permanent_data2/07_scRNA_groups/window_outputs_deep_cds_only'
```

```python
all_windows = glob.glob(os.path.join(input_dir, '*.exons.merged.txt'))
print(len(all_windows))
all_windows
```

```python
merged = pd.DataFrame()
progress = tnrange(len(all_windows))
for window in all_windows:
    df = pd.read_csv(window, sep='\t', index_col='name')[['edited_over_all_c']]
    df.columns = [
        os.path.basename(window).replace(
            '.fx.annotated.exons.merged.txt',''
        )
    ]
    merged = pd.merge(merged, df, how='outer', left_index=True, right_index=True)
    progress.update(1)
merged.fillna(0, inplace=True)
```

```python
merged.index.rename('Geneid', inplace=True)
merged = merged[sorted(merged.columns)]
merged.to_csv(os.path.join(output_dir, 'scored_edits_cds_only.txt'), sep='\t')
```
