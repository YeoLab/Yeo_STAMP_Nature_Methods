```r
for (dist_measures in all_dist_measures) {
    dist <- read.delim (dist_measures, header = T)

    dim(dist)
    print(paste0("DIST PRE TX DEDUP FOR ", dist_measures))
    print(dim(dist))
    # Determine longest length transcript for each gene.
    
    # I guess we don't really need this since we're already selecting one transcript per gene in notebook 02.
    # However, notebook 02 assumes hg38-gene annotations, and keeping this in will de-dup transcripts based on 
    # hg19-gene annotations, which we will be carrying forward. This results in a very minor reduction in our 
    # edit list (ie. both ENST00000409145.1 and ENST00000410042.1 belong to MRPL30 in hg19 but refer to 
    # different genes MRPL30-005 and RP11-38C17.1-001 in hg38). Not a big difference either way, but it's 
    # worth noting since the numbers don't 100% match up. 
    
    trx_len <- dist$utr5_size + dist$cds_size + dist$utr3_size # Determine transcript length
    temp <- data.frame(dist$gene_name, dist$refseqID, trx_len)
    colnames(temp) <- c("gene_name", "gid", "trx_len") 
    temp <- temp[order(temp$gene_name, -temp$trx_len),]
    temp <- temp[!duplicated(temp$gene_name),]
    # limit m6a data to one transcript per gene (longest)
    dist <- dist[dist$refseqID %in% temp$gid,]
    print(paste0("DIST POST TX DEDUP FOR ", dist_measures))
    print(dim(dist))
    # View size of our dataset (rows, columns)
    dim(dist)

    qplot(dist$rel_location, geom="histogram") + geom_vline(xintercept = 1:2, col = "grey") + theme_bw()
    summary(data.frame(dist$utr5_size, dist$cds_size, dist$utr3_size))

    utr5.SF <- median(dist$utr5_size, na.rm = T)/median(dist$cds_size, na.rm = T)
    utr3.SF <- median(dist$utr3_size, na.rm = T)/median(dist$cds_size, na.rm = T)

    # assign the regions to new dataframes
    utr5.dist <- dist[dist$rel_location < 1, ]
    cds.dist <- dist [dist$rel_location < 2 & dist$rel_location >= 1, ]
    utr3.dist <- dist[dist$rel_location >= 2, ]

    utr5.dist$rel_location <- rescale(utr5.dist$rel_location, to = c(1-utr5.SF, 1), from = c(0,1))
    utr3.dist$rel_location <- rescale(utr3.dist$rel_location, to = c(2, 2+utr3.SF), from = c(2,3))

    # Combine and plot
    ## Histogram
    metagene.coord <- c(utr5.dist$rel_location, cds.dist$rel_location, utr3.dist$rel_location)
    p <- qplot(metagene.coord, geom="histogram") + geom_vline(xintercept = 1:2, col = "grey") + theme_bw()

    qplot(metagene.coord, geom="freqpoly") + geom_vline(xintercept = 1:2, col = "red") + theme_bw()

    df <- data.frame(metagene.coord, 'red')
    ggplot(df) + geom_density(aes(x = metagene.coord, colour = 'red')) + xlim(0, 3) + 
      theme_bw() + geom_vline(xintercept = 1:2, col = "grey")

    write.table(dist$rel_location, paste0(dist_measures, '.rel_location.txt'))
}
```
