```python
conf = 0.5
prefix = 'Q1'  # can be 'IRES', 'TORIN', 'Q1', or 'top200'
```

```python
all_beds = sorted(glob.glob(os.path.join(input_dir, '*.conf{}.bed'.format(conf))))
print(len(all_beds))
all_beds[:3]
```

```python
regions = os.path.join(input_dir, 'riboseq2.region_sizes.{}.txt'.format(prefix))
for bed in all_beds:
    # q1_bed = os.path.splitext(bed)[0] + ".annotated.GSE112353_Riboseq.{}.bed".format(prefix)
    q1_bed = os.path.splitext(bed)[0] + ".annotated.Riboseq2.{}.bed".format(prefix)
    # output_txt = os.path.join(output_dir, os.path.splitext(os.path.basename(bed))[0] + ".GSE112353_Riboseq.{}.dist.measures.txt".format(prefix))
    output_txt = os.path.join(output_dir, os.path.splitext(os.path.basename(bed))[0] + ".Riboseq2.{}.dist.measures.txt".format(prefix))
    cmd = 'module load bioperl;'
    cmd += 'perl metaPlotR/rel_and_abs_dist_calc.pl --bed {} --regions {} > {}'.format(
        q1_bed, 
        regions,
        output_txt
    )
    print("------------------------------------------------------------------------------------------------------")
    print(cmd)
    ! $cmd
    print("------------------------------------------------------------------------------------------------------")
```
