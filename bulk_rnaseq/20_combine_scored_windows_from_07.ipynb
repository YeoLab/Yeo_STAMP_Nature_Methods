# combine the 'new scores' 

```python
import pandas as pd
import numpy as np
import glob
import os
from tqdm import tnrange, tqdm_notebook
```

```python
input_dir = '/home/bay001/projects/kris_apobec_20200121/permanent_data/final_analysis/01_SAILOR_bulk_rnaseq/window_outputs/'
output_dir = '/home/bay001/projects/kris_apobec_20200121/permanent_data/final_analysis/01_SAILOR_bulk_rnaseq/window_outputs/'
```

```python
all_windows = glob.glob(os.path.join(input_dir, '*.exons.merged.txt'))
print(len(all_windows))
all_windows[:3]
```

```python
merged = pd.DataFrame()
progress = tnrange(len(all_windows))
for window in all_windows:
    df = pd.read_csv(window, sep='\t', index_col='name')[['edited_over_all_c']]
    df.columns = [
        os.path.basename(window).replace(
            '.fastqTr.sorted.STARUnmapped.out.sorted.STARAligned.out.sorted', '').replace(
            '_a0_b0_e0.01.annotated.exons.merged.txt',''
        )
    ]
    merged = pd.merge(merged, df, how='outer', left_index=True, right_index=True)
    progress.update(1)
merged.fillna(0, inplace=True)
```

```python
merged.index.rename('Geneid', inplace=True)
merged = merged[sorted(merged.columns)]
merged.to_csv(os.path.join(output_dir, 'scored_edits.txt'), sep='\t')
```

# Now combine the edit/C windows across 5'UTR and CDS regions (skipping 3'UTR)

```python
all_windows = glob.glob(os.path.join(input_dir, '*.exons_no3utr.merged.txt'))
print(len(all_windows))
all_windows[:3]
```

```python
merged = pd.DataFrame()
progress = tnrange(len(all_windows))
for window in all_windows:
    df = pd.read_csv(window, sep='\t', index_col='name')[['edited_over_all_c']]
    df.columns = [os.path.basename(window).replace('.fastqTr.sorted.STARUnmapped.out.sorted.STARAligned.out.sorted_a0_b0_e0.01.annotated.exons_no3utr.merged.txt','')]
    merged = pd.merge(merged, df, how='outer', left_index=True, right_index=True)
    progress.update(1)
merged.fillna(0, inplace=True)
```
