```python
# Useful for normalizing/scaling across the length of CDS instead of gene length (a la RPKM).
```

```python
import pandas as pd
import numpy as np
import os
import glob
```

```python
input_bed = '/projects/ps-yeolab3/bay001/annotations/hg19/gencode_v19/hg19_v19_cds.bed'
output_lengths = '/home/bay001/projects/kris_apobec_20200121/permanent_data/final_analysis/01_SAILOR_bulk_rnaseq/outputs/hg19_v19_cds_lengths.txt'
```

```python
cds = pd.read_csv(input_bed, sep='\t', names=['chrom','start','end','name','score','strand'])
cds.head()
```

```python
cds['length'] = cds['end'] - cds['start']
cds.head(20)
```

```python
tot_cds_length = pd.DataFrame(cds.groupby('name')['length'].sum())
tot_cds_length.head()
```

```python
tot_cds_length.to_csv(output_lengths, sep='\t', index=True, header=True)
```
